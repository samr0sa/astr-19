```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating 1000 random numbers from my chosen distribution
# Here, I'm using the exponential distribution
random_numbers = np.random.exponential(1, 1000)


plt.hist(random_numbers, bins=100)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers (Exponential Distribution)')

plt.savefig('random_numbers_histogram2.pdf')

plt.show()
```
