```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.rand(1000)
```

```python
plt.hist(random_numbers, bins=100)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers')
plt.savefig('random_numbers_histogram.pdf')
plt.show()
```
