```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating 1000 random numbers distributed uniformly between [0, 1]
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100)

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.title('Histogram of 1000 Random Numbers')

plt.savefig('random_numbers_histogram.pdf')
plt.show()
```
