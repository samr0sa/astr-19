```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the array, x ranging from 0 to 1 in 100 steps

x = np.linspace(0, 1, 100)

def exp(x):
    return np.exp(x)
```

```python
# Calculate y using the compute_y function
y = exp(x)

# Create the plot
plt.plot(x, y)

# Label axes
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Awesomeness vs. Time")
plt.grid(True)

# Save figure as a PDF
plt.savefig("awesomeness_plot.pdf")
plt.show()

```
